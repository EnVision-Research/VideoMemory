```python
import os, json

from dotenv import load_dotenv
load_dotenv()

from langchain_core.messages import SystemMessage, HumanMessage
from langchain.agents import create_agent

from src.llm import init_llm_model
from src.schema import Script, Storyboard, Keyframe
from src.tools.nano_banana import nano_banana_replicate_tool
from src.prompts import SCREENWRITER, STORYBOARD, KEYFRAME

```

```python
# Load storyboard
thread_id = "02b5425c-e76e-4418-beeb-61e8ff37c425"
base_path = f"output/{thread_id}"
storyboard_path = f"{base_path}/storyboard.json"

with open(storyboard_path, "r") as f:
    storyboard = f.read()
```

```python

keyframe_agent = create_agent(
    model="deepseek:deepseek-chat",
    tools=[nano_banana_replicate_tool],
    system_prompt=KEYFRAME,
)

query = f"""
The base path is:\n {base_path}.

The storyboard is:\n {storyboard}.
"""

for msgs in keyframe_agent.stream(
    input={"messages": [("user", query)]}, 
    stream_mode="values", 
    config={"recursion_limit": 100}
):
    msgs['messages'][-1].pretty_print()
```
